```python
#1.import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import  train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
#2.Load Data
df=pd.read_csv("//content/Telco_Customer_Churn.csv")
```

```python
df.head()
```

```python
df.isnull().sum().astype(int)
```

```python
(df== " ").sum()
```

```python
#3.Special Placeholders ("?", "NA", "null")
df.isin(["?", "NA", "N/A", "null", "NULL"]).sum()
```

```python
# 4.convert blanks to numeric
df['TotalCharges']= pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
#5.Drop
df = df.dropna(subset=['TotalCharges']).reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
#Drop unwanted columns like customer id
df=df.drop(columns=['customerID'])
```

```python
#6.Split features and Target
x = df.drop("Churn", axis=1)
y=df['Churn']
```

```python
#7.Encode Categorical Variabels(features
binary_cols = ['gender', 'Partner', 'PhoneService']
x[binary_cols] = x[binary_cols].apply(LabelEncoder().fit_transform)
```

```python
#8.one hot encode for multi features
x=pd.get_dummies(x, drop_first=True)
```

```python
#9.Encode Target Column
y=y.map({"Yes":1, "No":0}).astype(int)
```

```python
#10. train_test_split(avoid leakage byscaling after split
x_train, x_test, y_train, y_test=train_test_split(
    x,y, test_size=0.2, random_state=42, stratify=y
)
```

```python
#feature scaling(fit on train, transform test)
scaler=StandardScaler()
x_train= pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns, index=x_train.index)
x_test=pd.DataFrame(scaler.transform(x_test), columns=x_test.columns, index=x_test.index)
```

```python
#Train a model
# Train Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train, y_train)

# Train Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)

# ==== Predict ====
# Logistic Regression Predictions
y_pred_lr  = lr.predict(x_test)
y_proba_lr = lr.predict_proba(x_test)[:, 1]

# Random Forest Predictions
y_pred_rf  = rf.predict(x_test)
y_proba_rf = rf.predict_proba(x_test)[:, 1]
```

```python
print("Logistic Regression:")
print("  Accuracy :", accuracy_score(y_test, y_pred_lr))
```
